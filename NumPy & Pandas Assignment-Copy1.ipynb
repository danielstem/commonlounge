```python
for col in ['BloodPressure','SkinThickness','Insulin','BMI','DiabetesPedigreeFunction']:
    temp = (dataset[col]==0)
    dataset.loc[temp,col]=None

```

```python
normalized_dataset= (dataset-dataset.mean())/dataset.std()
```

```python
normalized_dataset.mean()
```

```python
normalized_dataset.std()
```

```python
normalized_dataset.iloc[2]
```

```python
cols=list(normalized_dataset.columns)[:-1]
for column in cols:
    temp=normalized_dataset[column].isnull()
    normalized_dataset.insert(0,"%s_isNull" % column,temp)
normalized_dataset[normalized_dataset.isnull()]=0.0
normalized_dataset = normalized_dataset.applymap(float)
```
